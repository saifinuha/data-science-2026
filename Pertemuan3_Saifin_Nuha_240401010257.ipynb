```python
# ============================================
# PIPELINE DATA CLEANING — HOUSING DATASET
# ============================================
import pandas as pd, numpy as np
from scipy.stats.mstats import winsorize
# STEP 0 — Load & eksplorasi awal
# Upload file di colab
# df = pd.read_csv('housing_dirty.csv')
# Upload di github
df = pd.read_csv('https://raw.githubusercontent.com/saifinuha/data-science-2026/main/housing_dirty.csv')
print('Shape awal:', df.shape)
print(df.isnull().sum())

# STEP 1 — Hapus Duplikat
df.drop_duplicates(inplace=True)
print('Setelah hapus duplikat:', df.shape)

# STEP 2 — Normalisasi String
df['kota'] = df['kota'].str.strip().str.title()
df['kondisi'] = df['kondisi'].str.strip().str.lower()

# STEP 3 — Imputasi Missing Values
df['luas_m2'] = df['luas_m2'].fillna(df['luas_m2'].median())
df['harga_juta'] = df['harga_juta'].fillna(df['harga_juta'].median())
df['kamar'] = df['kamar'].fillna(df['kamar'].mode()[0])

# STEP 4 — Tangani Outlier (IQR Fence)
for col in ['harga_juta', 'luas_m2', 'tahun_bangun']:
  Q1, Q3 = df[col].quantile([0.25, 0.75])
  IQR = Q3 - Q1
  df[col] = df[col].clip(Q1-1.5*IQR, Q3+1.5*IQR)

# STEP 5 — Validasi & Ekspor
assert df.isnull().sum().sum() == 0, 'Masih ada missing!'
assert df.duplicated().sum() == 0, 'Masih ada duplikat!'

print('Shape akhir:', df.shape)
df.to_csv('housing_clean.csv', index=False)
print('Dataset bersih tersimpan!')
```

### Apa yang dapat dipelajari dari pipeline ini:

Pipeline ini menunjukkan alur kerja standar untuk membersihkan dataset perumahan, meliputi beberapa langkah pra-pemrosesan data yang esensial:

1.  **Pemuatan Data dan Eksplorasi Awal:**
    *   Cara memuat data dari file CSV, baik secara lokal maupun dari URL (misalnya, GitHub).
    *   Pemeriksaan awal untuk bentuk dataset (`df.shape`) dan nilai yang hilang (`df.isnull().sum()`).

2.  **Penanganan Data Duplikat:**
    *   Menggunakan `df.drop_duplicates(inplace=True)` untuk menghapus baris redundan dari dataset.

3.  **Normalisasi String:**
    *   Membersihkan data teks menggunakan `.str.strip()` untuk menghapus spasi di awal/akhir.
    *   Menstandardisasi huruf dengan `.str.title()` (untuk 'kota') dan `.str.lower()` (untuk 'kondisi') untuk memastikan konsistensi.

4.  **Imputasi Nilai yang Hilang:**
    *   Menangani nilai numerik yang hilang menggunakan `fillna()` dengan nilai median (`df.median()`). Ini adalah pendekatan yang kuat untuk menghindari distorsi dari outlier.
    *   Menangani nilai kategorikal yang hilang menggunakan `fillna()` dengan nilai modus (`df.mode()[0]`).

5.  **Penanganan Outlier (Metode Pagar IQR):**
    *   Mengidentifikasi dan membatasi outlier untuk kolom numerik (`harga_juta`, `luas_m2`, `tahun_bangun`) menggunakan metode Interquartile Range (IQR).
    *   Menggunakan `df.clip()` untuk membatasi nilai dalam batas bawah dan atas yang dihitung, secara efektif menangani nilai ekstrem tanpa menghapus seluruh baris.

6.  **Validasi Data dan Ekspor:**
    *   Menggunakan pernyataan `assert` untuk secara terprogram memverifikasi bahwa langkah-langkah pembersihan data berhasil (misalnya, tidak ada nilai yang hilang, tidak ada duplikat yang tersisa).
    *   Mengekspor DataFrame yang sudah bersih ke file CSV baru (`housing_clean.csv`) tanpa indeks.
